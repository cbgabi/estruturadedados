```python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "nota": [8.5, 6.0, 7.2, 5.5, 9.0],
    "faltas": [2, 6, 4, 8, 1]
}

df = pd.DataFrame(dados)
mais_5_faltas = df[df["faltas"] > 5]
print("\nAlunos com mais de 5 faltas:")
print(mais_5_faltas)
```

```python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "nota": [8.5, 6.0, 7.2, 5.5, 9.0],
    "faltas": [2, 6, 4, 8, 1]
}

df = pd.DataFrame(dados)
ordenado = df.sort_values(by="nota", ascending=False)
print("\nAlunos ordenados por nota (decrescente):")
print(ordenado)
```

```python
import pandas as pd
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "nota": [8.5, 6.0, 7.2, 5.5, 9.0],
    "faltas": [2, 6, 4, 8, 1]
}

df = pd.DataFrame(dados)
df["situacao"] = df["nota"].apply(lambda x: "Aprovado" if x >= 7 else "Reprovado")
print("\nDataFrame com situação:")
print(df)
```
